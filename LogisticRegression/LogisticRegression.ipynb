```python
class Crossentropy():
    '''inputs:
    X: (n, m)
    y: (n, 1) 
    w: (m, 1) - weights
    j: index'''
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def reg(self, w):
        '''return weights for the regularization term
        i.e. weights with intercept weight = 0 because we do not regularize intercept'''
        w_reg = w.copy()
        w_reg[0] = 0
        return w_reg
    
    def natural_log(self, x):
        '''input: x is numpy array
        take the natual log, replaces log(0) by log of a small number'''
        x[x==0] = 1e-20
        return np.log(x)
    
    def compute_cost(self, X, w, y, reg_lambda):
        y_hat = self.sigmoid(np.matmul(X, w))
        reg_term = reg_lambda * self.reg(w)**2
        self.cost = (- np.sum( y * self.natural_log(y_hat) + (1 - y) * self.natural_log(1 - y_hat) ) + np.sum(reg_term) ) / y.shape[0]
    
    def derivative(self, X, w, y, reg_lambda):
        y_hat = self.sigmoid(np.matmul(X, w))
        reg_term = reg_lambda * self.reg(w)
        return ( np.sum((y_hat - y) * X, axis=0).reshape(-1,1) + reg_term ) / y.shape[0]
    
class LogisticRegression():
    
    def __init__(self, learning_rate = 0.001, max_iter = 100, threshold = 1e-3, reg_lambda = 0, debug = False, seed = None):
        self.coef = {} 
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.threshold = threshold
        self.reg_lambda = reg_lambda
        self.debug = debug
        self.seed = seed
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def add_intercept(self, X):
        X_zero = np.ones((X.shape[0],1))
        return np.concatenate((X_zero, X), axis=1)
    
    def to_binary(self, y, i):
        '''input: i = the class label to consider as 1
        turns multiclass y into binary y'''
        y = y + 2 # in case i in [0,1]
        y[y != i+2] = 0
        y[y == i+2] = 1
        return y
    
    def fit_multiclass(self, X, y, batch_size):
        y = y.reshape(y.shape[0], 1)
        # add intercept term
        X = self.add_intercept(X)
        # init weights
        init_w = np.random.randn(X.shape[1],1)/ 2
        # for each class in y, create binary y, fit regression and store coefficients
        classes = np.unique(y)
        for cl in classes:
            y_binary = self.to_binary(y, cl)
            self.fit_binary(X, y_binary, cl, batch_size)
             
    def fit_binary(self, X, y, target_class, batch_size):
        y = y.reshape(y.shape[0], 1)
        # add intercept term
        X = self.add_intercept(X)
        # init weights
        init_w = np.random.randn(X.shape[1],1)/ 2
        # gradient descent with Crossentropy loss
        gr = GradientDescent(X, y, init_w, loss = Crossentropy(), batch_size = batch_size, reg_lambda = self.reg_lambda, seed = self.seed)
        gr.gradient_descent(learning_rate = self.learning_rate, 
                            max_iter = self.max_iter, 
                            threshold = self.threshold, 
                            debug = self.debug)
        self.coef[target_class] = gr.w
        
    def fit(self, X, y, batch_size):
        if len(np.unique(y)) > 2:
            self.fit_multiclass(X, y, batch_size)
        else:
            target_class = max(y)
            self.fit_binary(X, y, target_class, batch_size)
             
    def predict_binary(self, X_test, target_class, pred_threshold, pred_prob):
        # add intercept term
        X_test = self.add_intercept(X_test)            
        y_pred = self.sigmoid(np.matmul(X_test, self.coef[target_class])).reshape(X_test.shape[0],)
        if not pred_prob:
            y_pred[y_pred < pred_threshold] = 0
            y_pred[y_pred >= pred_threshold] = 1
        return y_pred # shape = kx1
    
    def predict_multiclass(self, X_test, pred_threshold):
        # add intercept term
        X_test = self.add_intercept(X_test)
        # one prediction array per class
        y_pred = []
        classes = list(self.coef.keys())
        for cl in classes:
            y_pred.append((self.predict_binary(X_test, target_class = cl, pred_threshold = 0.5, pred_prob = True)).reshape(-1,1))
        y_pred = np.concatenate(tuple(y_pred), axis=1)
        y_pred = np.argmax(y_pred, axis=1)
        return y_pred # shape = kx1
    
    def predict(self, X_test, pred_threshold = 0.5, pred_prob = False):
        '''input:
        X_test: (k, m)'''
        if len(self.coef) == 1:
            target_class = list(self.coef.keys())[0]
            return self.predict_binary(X_test, target_class, pred_threshold, pred_prob)
        else:
            return self.predict_multiclass(X_test, pred_threshold)
```
