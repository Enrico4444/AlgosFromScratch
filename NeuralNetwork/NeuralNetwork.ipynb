```python
def one_hot(y):
    # credit: https://stackoverflow.com/questions/29831489/convert-array-of-indices-to-1-hot-encoded-numpy-array
    '''y must be integer encoded. 
    All integers smaller or equal than max value of y are assumed to exist in the population'''
    y = y.reshape(-1,)
    placeholder = np.zeros((y.shape[0], int(y.max()+1)))
    placeholder[np.arange(y.size),y] = 1
    return placeholder
```

## Algorithm

```python
class NeuralNetwork():
    def __init__(self, hidden_layers, output_layers, activations, seed = None):
        '''hidden_layers and output_layers: list of int, 
        each element is the number of neurons in the corresponding layer
        activation: list of string with name of activation for each layer'''
        self.layers = hidden_layers + output_layers
        self.cost = 0
        self.history = []
        self.activation_functions = {'Sigmoid':self.Sigmoid(), 'Linear':self.Linear(), 'Softmax':self.Softmax()}
        self.activations = [self.activation_functions[i] for i in activations]
        self.losses = {'MSE':self.MSE(), 'Crossentropy':self.Crossentropy()}
        self.seed = seed
         
    def init_network(self):
        '''randomly init weights and biases, init activations to zero'''
        self.w = [] # weights
        self.b = [] # biases
        self.z = [] # z
        self.a = [] # activations
        for l in range(len(self.layers)):
            if l == 0:
                n_neurons_prev = self.X.shape[1]
            else:
                n_neurons_prev = self.layers[l-1]
            n_neurons = self.layers[l]
            np.random.seed(self.seed)
            self.w.append(np.random.randn(n_neurons, n_neurons_prev)/ 2) 
            self.b.append(np.random.randn(n_neurons, 1)/ 2) 
            self.z.append(np.zeros((n_neurons, 1)))
            self.a.append(np.zeros((n_neurons, 1)))
    
    def forward_pass(self, x_batch):
        for l in range(len(self.layers)):
            if l == 0:
                x = x_batch.T
            else:
                x = self.a[l-1]
            self.z[l] = np.matmul(self.w[l], x) + self.b[l]
            self.a[l] = self.activations[l].activation(self.z[l])
                
    def backpropagation(self, x_batch, y_batch, loss):
        x = x_batch.T
        y = y_batch.T
        # C_wL, C_bL
        C_aL = self.losses[loss].der(self.a[-1], y)
        aL_zL = self.activations[-1].der(self.z[-1])
        zL_wL = self.a[-2]
        temp = C_aL * aL_zL
        self.db[-1] += np.sum(temp, axis=1, keepdims = True)
        self.dw[-1] += np.matmul(temp, zL_wL.T)        
        # C_wl, C_bl
        gr = C_aL
        reverse_layers = list(range(len(self.layers)-1))[::-1]
        for l in reverse_layers:
            aL_zL = self.activations[l+1].der(self.z[l+1])
            zL_al = self.w[l+1]
            al_zl = self.activations[l].der(self.z[l])
            if l == 0:
                zl_wl = x
            else:
                zl_wl = self.a[l-1].T
            gr = gr * aL_zL
            gr = np.matmul(gr.T, zL_al).T
            temp = gr * al_zl
            self.db[l] += np.sum(temp, axis=1, keepdims = True)
            self.dw[l] += np.matmul(temp, zl_wl.T)
            
    def param_update(self, learning_rate, n):
        self.dw = [w / n for w in self.dw]
        self.db = [b / n for b in self.db]
        self.w = [self.w[i] - learning_rate * self.dw[i] for i in range(len(self.w))]
        self.b = [self.b[i] - learning_rate * self.db[i] for i in range(len(self.b))]
                
    def reset_deltas(self):
        self.dw = []
        self.db = []
        for w,b in zip(self.w, self.b):
            self.dw.append(np.zeros(w.shape))
            self.db.append(np.zeros(b.shape))
            
    def fit(self, X, y):
        self.X = X
        if len(y.shape) == 1:
            y = y.reshape(y.shape[0], 1) # if y has shape (m,) reshape to (m,1)
        self.y = y
        self.init_network()
        self.reset_deltas()
        
    def split_data(self, batch_size):
        data = np.concatenate((self.X, self.y), axis=1)
        np.random.shuffle(data)
        self.X = data[:, :-self.y.shape[1]]
        self.y = data[:, -self.y.shape[1]:]
        if len(self.y.shape) == 1:
            self.y = self.y.reshape(self.y.shape[0], 1)
        q = self.X.shape[0] // batch_size
        block_end = q * batch_size
        X_batches = np.split(self.X[:block_end], q) 
        y_batches = np.split(self.y[:block_end], q) 
        if self.X.shape[0] % batch_size != 0:
            X_batches = X_batches + [self.X[block_end:]] 
            y_batches = y_batches + [self.y[block_end:]]
        return X_batches, y_batches
        
    def train(self, loss, learning_rate = 0.001, n_epochs = 100, batch_size = 1):
        x_batches, y_batches = self.split_data(batch_size)
        for epoch in range(n_epochs):
            for x_batch, y_batch in zip(x_batches, y_batches):
                n = x_batch.shape[0]
                self.forward_pass(x_batch)
                self.backpropagation(x_batch, y_batch, loss)
                self.param_update(learning_rate = learning_rate, n=n)
                self.reset_deltas()
                # add the cost of this batch
                self.cost += self.compute_loss(loss, self.a[-1], y_batch.T)
            # at the end of the epoch, calculate average (over training example and training classes) epoch cost and add to history, then reset cost
            self.history.append(self.cost / self.y.shape[0] / self.y.shape[1])
            self.cost = 0
            if epoch % int(n_epochs/10) ==0 and epoch != 0:
                print('Epoch {}, loss {}'.format(epoch, round(self.history[-1], 4)))
        print('Loss {}'.format(round(self.history[-1], 4)))
                
    def compute_loss(self, loss, y_hat, y):
        return self.losses[loss].cost(y_hat, y)
    
    def format_multiclass(self, y_pred):
        placeholder = np.zeros((y_pred.shape))
        placeholder[np.arange(y_pred.shape[0]),np.argmax(y_pred, axis=1)] = 1
        return placeholder
                
    def predict(self, X_test, pred_prob = True, threshold = 0.5, multilabel = False):
        '''returns y_pred in the original input shape of (n_samples, n_classes)'''
        a = [X_test.T]
        for l in range(len(self.layers)):
            z = np.matmul(self.w[l], a[l]) + self.b[l]
            a.append(self.activations[l].activation(z))
        y_pred = a[-1].T
        if not pred_prob:
            if y_pred.shape[1] > 1 and not multilabel: # multiclass
                y_pred = self.format_multiclass(y_pred)
            else:
                y_pred = (y_pred>=threshold).astype(int)
        return y_pred
    
    ##### losses 
    class MSE():
        def cost(self, y_hat, y):
            return np.sum((y_hat - y)**2) 
        def der(self, y_hat, y):
            y = y.reshape(y_hat.shape)
            return 2*(y_hat - y) # average over all classes
    
    class Crossentropy():
        def natural_log(self, x):
            '''input: x is numpy array
            take the natual log, replaces log(0) by log of a small number'''
            x[x==0] = 1e-20
            return np.log(x)
        def cost(self, y_hat, y):
            return - np.sum(y * self.natural_log(y_hat) + (1 - y) * self.natural_log(1 - y_hat))
        def der(self, y_hat, y):
            y = y.reshape(y_hat.shape)
            return (y_hat - y) / (y_hat * (1 - y_hat))
        
    #### activations       
    class Sigmoid():
        def activation(self, z):
            return 1 / (1 + np.exp(-z))
        def der(self, z):
            return self.activation(z) * (1-self.activation(z))
        
    class Linear():
        def activation(self, z):
            return z
        def der(self, z):
            return np.ones(z.shape)
            
    
```
